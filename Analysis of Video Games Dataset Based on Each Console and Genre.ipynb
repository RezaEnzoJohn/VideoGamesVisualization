```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("vgsales.csv")
df.head(15)
df.info()
```

```python
#View the relationship between Genre and Platform
platformGenre = pd.crosstab(df.Platform,df.Genre)
platGenreTotal = platGenre.sum(axis=1).sort_values(ascending = False)
plt.figure(figsize=(12,8))
sns.barplot(x = platGenreTotal.index, y = platGenreTotal.values, orient='v')
plt.xlabel = "Platform"
plt.ylabel = "The amount of games"
plt.show()
platGenre.sum(axis=1).sort_values(ascending = False)
```

```python
platformGenre['Total'] = platformGenre.sum(axis=1)
PPlatform = platformGenre[platformGenre['Total']>1000].sort_values(by='Total', ascending = False)
Ddata = PPlatform.loc[:,:'Strategy']
maxi = Ddata.values.max()
mini = Ddata.values.min()
PPlatformfinal = PPlatform.append(pd.DataFrame(PPlatform.sum(), columns=['total']).T, ignore_index=False)
sns.set(font_scale=0.7)
plt.figure(figsize=(12,8))
sns.heatmap(PPlatformfinal, vmin = mini, vmax = maxi, annot=True, fmt="d", cmap="YlGnBu")
plt.xticks(rotation = 45)
plt.yticks(rotation = 30)
plt.show()
```

```python
GenreGroup = df.groupby(['Genre']).sum().loc[:, 'NA_Sales':'Global_Sales']
GenreGroup['NA_Sales%'] = GenreGroup['NA_Sales']/GenreGroup['Global_Sales']
GenreGroup['EU_Sales%'] = GenreGroup['EU_Sales']/GenreGroup['Global_Sales']
GenreGroup['JP_Sales%'] = GenreGroup['JP_Sales']/GenreGroup['Global_Sales']
GenreGroup['Other_Sales%'] = GenreGroup['Other_Sales']/GenreGroup['Global_Sales']
plt.figure(figsize=(12, 14))
sns.set(font_scale=0.7)
plt.subplot(211)
sns.heatmap(GenreGroup.loc[:, 'NA_Sales':'Other_Sales'], annot=True, fmt = '.1f', cmap="YlGnBu")
plt.title("Comparation of each area in each Genre")
plt.subplot(212)
sns.heatmap(GenreGroup.loc[:,'NA_Sales%':'Other_Sales%'], vmax =1, vmin=0, annot=True, fmt = '.2%', cmap="YlGnBu")
plt.title("Comparation of each area in each Genre(Pencentage)")
plt.show()
df.groupby(['Genre']).sum().loc[:, 'NA_Sales':'Global_Sales']
```
